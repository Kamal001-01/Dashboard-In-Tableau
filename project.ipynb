```python
df.rename(columns={'timestamp':'time','cnt':'count','t1':'temp_real_C','t2':'temp_feel_C','hum':'humidity_percent',
                    'wind_speed':'wind_kph','weather_code':'weather'},inplace=True)
```

```python
df
```

```python
# maping the season and weather
season_dict = {0.0:'Spring',    
               1.0:'Summer',   
               3.0:'Autumn',    
               2.0:'Winter'}
weather_dict = {1.0:'Clear',    
                2.0:'Scattered Clouds',     
                3.0:'Broken Clouds',     
                7.0:'Cloudy',    
                4.0:'Rain',     
                26.0:'Rain with Thunderstrom',     
                10.0:'Snowfall'}
```

```python
df['season']=df['season'].map(season_dict)
```

```python
df
```

```python
df['weather']=df['weather'].map(weather_dict)
```

```python
df.to_excel('london_bike_final.xlsx',index=False,sheet_name = 'Data')
```
